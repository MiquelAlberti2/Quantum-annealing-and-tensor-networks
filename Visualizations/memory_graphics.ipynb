```python
import numpy as np
import matplotlib.pyplot as plt



# Generate x values
x = np.linspace(-10, 15, 400)

# Calculate y values for the parabola
y_parabola = 0.0371 * x**2 - 0.90603 * x

# Calculate y values for the exponential function
y_exponential = np.exp(-x)

# Plot the parabola
plt.plot(x, y_parabola, label='h(x)')

# Plot the exponential function
plt.plot(x, y_exponential, label=r'$y = e^{-x}$')

# Add vertical line at x=0
plt.axvline(x=0, color='r', linestyle='--', label='x=0')

# Add legend
plt.legend()

# Add grid for better visualization
plt.grid(True)

# Add labels and title
plt.xlabel('x')
plt.ylabel('y')
plt.title('Parabola and Exponential Function Plot')

# Adjust x and y axis limits to zoom out a bit
plt.xlim(-5, 15)
plt.ylim(-2, 100)

# Show plot
plt.show()
```
